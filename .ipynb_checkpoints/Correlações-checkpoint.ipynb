# Matriz de Correlações por HeatMap

Dados a serem definidos

```python
# O LIMITE pode ser diferente para cada arquivo
# Os dados que usei so tinham observações de 2h, agrupado em 1min, portanto limite = 120 minutos
# caso fosse dados de UM dia agrupados em 1min, limite = 1440 (minutos num dia)
# caso fosse dados de UM dia agrupados em min, limite = 1440/5 = 288
limite = 120
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path3 = r'csv_radares-20200426T211130Z-001_selec_group_n\\' ## mudar aqui o caminho e nome do arquivo
                                                                        
##########################################  local ###########################                                                                                                                                           
a = pd.read_csv(path3 +'\\01\\' + "01_group_" + "1Min" + ".csv", index_col=["Local",'Especie','Classe'])
b = pd.read_csv(path3 +'\\01\\' + "01_group_" + "2Min" + ".csv")
c = pd.read_csv(path3 +'\\01\\' + "01_group_" + "3Min" + ".csv")
d = pd.read_csv(path3 +'\\01\\' + "01_group_" + "4Min" + ".csv")
e = pd.read_csv(path3 +'\\01\\' + "01_group_" + "5Min" + ".csv")
display(a)
```

```python
Radares = [5323,5226,5358,5357,5355,5356,5151,5205,5204,5313,5316,5315,5314,5476,5475,5474,5479,5478,5477,5480,5196,5195,5194,5197,5203,5202,5432,5354,5414,5413,5433,5098,5496,5129,5471,5470,5199,5198,5201,5200,5508,5507,5308,5307]

Radares = list(dict.fromkeys(Radares)) # Para tirar duplicados

velocidades = []

for n in Radares:
    #print(n)

    # Não são todos os IDs de radares que estão presentes em todos os arquivos excel
    # Portanto, os subsets de "Locais" que não estao no set são preenchidos com 0
    # foi feito assim para poder usar o mesmo codigo em todos os arquivos, 
    # já que uns tem tantos radares que outros nao tem
    
    
    try: 

        a_subset = a.xs((n,0,0), axis = 0) # n é o LOCAL em questao, 0 é so pra pegar "Carros"
    
    except:
    
    
        # aqui, quando o radar nao é encontrado, já criamos um subset com zeros
        a_subset = pd.DataFrame({"Local":0000, "Data":a_subset['Data'], "Vel_media":0, "Vel_mediana":0, "Vel_std":0, "Volume":1})

    # para a correlação, os dados tem que estar no seguinte formato:
    # Cada linha é uma variável, no nosso caso, cada radar
    # Cada coluna é uma observação, é cada uma das Vel_media, Vel_std, Vel_mediana
    # Para que a correlação seja valida, cada radar deve ter o mesmo numero de observações
    # Sendo assim, fazemos o preenchimento do DF com a ULTIMA OBSERVAÇÃO


    # O LIMITE pode ser diferente para cada arquivo
    # Os dados que usei so tinham observações de 2h, agrupado em 1min, portanto limite = 120 minutos
    # caso fosse dados de UM dia agrupados em 1min, limite = 1440 (minutos num dia)
    # caso fosse dados de UM dia agrupados em min, limite = 1440/5 = 288
    
    while a_subset.shape[0] < limite : #faz preenchimento com o ulitmo para a correlação
        a_subset = a_subset.append(a_subset.tail(1),ignore_index=True) 

    #display(a_subset)      #para ver cada subset preenchido 



    # coloca tudo no velocidades
    velocidades.append(a_subset['Vel_media'].values)



#########################################################################################
# Apenas pra saber se o preenchimento e tamanho do "velocidades" está certo
# Se estiver correto, nos fomos assertivos na validação

for i in range(len(velocidades)):
    assert len(velocidades[i]) == limite


#print(len(velocidades))
```

```python
corr = np.corrcoef(velocidades)

# necessario para o posicionamento correto das labels
posições = np.arange(1,len(velocidades)+1)-0.5

# usado para mudar o tamanho da figura
plt.figure(figsize=(8,8))

ax = sns.heatmap(corr, linewidths=.5) #linewidths para aumentar os espaços brancos
plt.yticks(posições,Radares,rotation = 0)
plt.xticks(posições,Radares,rotation = 90)

# as linhas e colunas brancas são os RADARES QUE NÃO ESTAO PRESENTES NO DATASET (ZEROS)

plt.show()
```
