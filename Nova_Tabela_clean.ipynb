```python
import pandas as pd
import numpy as np
from datetime import timedelta
from datetime import datetime

num_agrupado = ["10875","10862","10874","10858","10858","10893","10893","10864","10848","10906","10900","10904","10905","10605","10433","10426","10468","10508","10508","10613","10538","10616","10631","10505","10533","10607","10524","10479","10565","10586","10583","10587","10587","10584","10584","10514","10514","10513","10513","10595","10595","10594","10594","10582","10582","10585","10585","10555","10552","10488","10615","10464","10560","10543","10543","10540","10542","10512","10509","10509","10438","10497","10532","10532","10488","10502","10459","10454","10480","10591","10540","10591","10520","10554","10590","10537","10609","10609","10506","10506","10490","10490","10557","10557","10517","10556","10556","10426","10426","10568","10596","10596","10433","10433","10433","10628","10622","10619","10613","10538","10616","10468","10476","10476","10276","10306","10305","10348","10348","10337","10337","10309","10309","10310","10310","10302","10302","10293","10352","10262","10262","10353","10235","10269","10291","10292","10434","10434","10408","10371","10368","10380","10381","10434","10381","10381","10933","10940","10881","10877","10947","10911","10933","10894","10895","10770","10725","10655","10655","10702","10702","10701","10701","10630","10630","10674","10675","10670","10656","10654","10654","10720","10692","10732","10684","10709","10672","10788","10715","10807","10807","10683","10683","10762","10762","10834","10834","10761","10761","10748","10748","10680","10680","10679","10679","10635","10635","10808","10658","10664","10653","10820","10815","10815","10724","10713","10712","10711","10810","10689","10634","10634","10632","10632","10745","10800","10800","10670","10689","10742","10742","10747","10747","10798","10798","10665","10665","10830","10667","10667","10652","10652","10669","10669","10698","10698","10697","10697","10640","10640","10686","10686","10736","10736","10731","10731","10789","10703","10829","10729","10656","10764","10678","10763","10637","10637","10637","10644","10644","10644","10644","10658","10767","10704","10673","10627","10627","10625","10625","10592","10592","10578","10578","10650","10527","10527","10534","10474","10474","10471","10471","10531","10531","10589","10511","10511","10521","10623","10564","10564","10614","10614","10577","10577","10536","10620","10620","10477","10498","10498","10526","10526","10588","10530","10536","10645","10325","10323","10507","10356","10484","10482","10277","10431","10478","10606","10606","10539","10539","10487","10452","10458","10475","10384","10374","10566","10398","10357","10304","10544","10544","10375","10440","10334","10363","10363","10344","10344","10431","10321","10321","10351","10317","10529","10541","10541","10528","10528","10563","10500","10410","10410","10451","10525","10457","10346","10346","10496","10496","10489","10489","10361","10361","10358","10358","10367","10367","10420","10420","10421","10421","10465","10465","10561","10377","10290","10314","10486","10413","10413","10356","10356","10484","10484","10482","10482","10482","10357","10492","10492","10516","10572","10572","10410","10420","10421","10377","10486","10413","10413","10374","10374"]


Radares = [5415,5483,5126,5471,5470,5316,5315,5313,5314,5358,5357,5355,5356,5463,5297,5296,5354,5098,5476,5475,5474,5479,5478,5477,5480,5323,5226,5308,5307,5432,5433,5414,5413,5201,5200,5199,5198,5205,5204,5203,5202,5196,5197,5195,5194,5445,5104,5446,5103,5462,5461,5460,5441,5106,5443,5442,5439,5105,5440,5457,5449,5317,5101,5506,5336,5335,5402,5401]


# cria um dicionário
def faz_dict (keys, values):
    '''
        Cria um dicionário com a correspondência entre 
        chaves e valores. Com a função .map() pode, ser
        adicionadas novas colunas no DF.

        Inputs:
            keys: (list) chaves que estão no presente DF
            values: (list) valores que serão inputados na nova coluna

        Return:
            mapa: (dict) mapa de correspondência
    '''

    mapa = {}
    for i in range(len(keys)):
        mapa[keys[i]] = values[i]

    return mapa
```
